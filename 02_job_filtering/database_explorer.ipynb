# Database Explorer
Explore the jobs_new.db database with tables, metadata, and data

```python
import sqlite3
import pandas as pd
from IPython.display import display, HTML

# Database path
db_path = "../jobs_new.db"
```

```python
# Connect to database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Get all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in database:")
for table in tables:
    print(f"  - {table[0]}")
```

```python
# Get metadata for each table
for table_name in [table[0] for table in tables]:
    print(f"\n{'='*50}")
    print(f"Table: {table_name}")
    print(f"{'='*50}")
    
    # Get table schema
    cursor.execute(f"PRAGMA table_info({table_name});")
    columns = cursor.fetchall()
    
    print("\nColumns:")
    for col in columns:
        print(f"  {col[1]} ({col[2]}) - NOT NULL: {col[3]} - DEFAULT: {col[4]}")
    
    # Get row count
    cursor.execute(f"SELECT COUNT(*) FROM {table_name};")
    row_count = cursor.fetchone()[0]
    print(f"\nTotal rows: {row_count}")
```

```python
# Display data as DataFrames for each table
for table_name in [table[0] for table in tables]:
    print(f"\n{'='*60}")
    print(f"DATA FOR TABLE: {table_name.upper()}")
    print(f"{'='*60}")
    
    # Load table into DataFrame
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    
    # Display basic info
    print(f"Shape: {df.shape} (rows, columns)")
    print(f"Columns: {list(df.columns)}")
    print("\nFirst 5 rows:")
    display(df.head())
    
    # Show data types
    print("\nData types:")
    print(df.dtypes)
    
    # Show basic statistics for numeric columns
    numeric_cols = df.select_dtypes(include=['number']).columns
    if len(numeric_cols) > 0:
        print("\nNumeric columns statistics:")
        display(df[numeric_cols].describe())
```

```python
# Display all data in a clean format
for table_name in [table[0] for table in tables]:
    print(f"\n{'='*80}")
    print(f"COMPLETE DATA FOR TABLE: {table_name.upper()}")
    print(f"{'='*80}")
    
    # Load table into DataFrame
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    
    # Display all rows with proper formatting
    if len(df) == 0:
        print("Table is empty")
    else:
        # Set display options to show all columns and rows
        pd.set_option('display.max_columns', None)
        pd.set_option('display.max_rows', None)
        pd.set_option('display.width', None)
        pd.set_option('display.max_colwidth', 50)
        
        display(df)
        
        # Show row and column separation
        print(f"\nRows: {df.shape[0]}, Columns: {df.shape[1]}")
        for i, col in enumerate(df.columns):
            print(f"Column {i+1}: {col}")
```

```python
# Specific query to see job postings with chances
print("Job Postings with Chances:")
df_jobs = pd.read_sql_query("""
    SELECT job_unique_id, job_title, company_name, experience_required, location, score
    FROM job_postings 
    WHERE score IS NOT NULL
    ORDER BY score DESC
""", conn)

display(df_jobs)

print(f"\nTotal jobs with scores: {len(df_jobs)}")
print(f"Score range: {df_jobs['score'].min()} - {df_jobs['score'].max()}")
```

```python
# Close connection
conn.close()
print("Database connection closed.")
```
